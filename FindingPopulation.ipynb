```python
cities = "Istanbul,Ankara,Izmir,Bursa"
population = "15000000,5500000,4300000,1800000" # csv (comma seperated values)

cityName = input("Please enter the city name: ").lower()
cityList = cities.lower().split(",")
populationList = population.split(",")

if cityName in cityList:
  position = cityList.index(cityName)
  print("Population of", cityName.capitalize(), "is", populationList[position])
else:
  print("The city name does not exist")
```
